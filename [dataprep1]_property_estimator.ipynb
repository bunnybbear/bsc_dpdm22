#เปรียบเทียบความแตกต่างในการประเมินราคาทรัพย์แต่ละทรัพย์ของนักประเมินแต่ละคนและของราคาตั้งขาย ใช้ MAPE เป็นเกณฑ์ในการเปรียบเทียบ

```python
from sklearn.metrics import mean_absolute_percentage_error as mape #mape
```

```python
from types import MethodType
```

##MAPE บ้าน

```python
dataH_P2['ประเมินโดย 1'] = dataH_P2['post id'].map(dataH_P1.set_index('post id')['ประเมินราคาซื้อ-ขาย'])
dataH_P2.head()
```

```python
dataH_P2.rename(columns={'ประเมินราคาซื้อ-ขาย' :'ประเมินโดย 2'}, inplace=True)
```

```python
MHP = dataH_P2[['post id','ราคาทรัพย์','ประเมินโดย 1','ประเมินโดย 2']]
MHP
```

```python
#เช็ค missing
MHP.isnull().any()
```

```python
#แก้ missing
MHP = MHP.fillna(value={'ประเมินโดย 2':dataH_P2['ประเมินโดย 2'].mean()})
```

```python
MHP.isnull().any()
```

```python
ATH = MHP[['ราคาทรัพย์']]
PH1 = MHP[['ประเมินโดย 1']]
PH2 = MHP[['ประเมินโดย 2']]
```

```python
MH1 = mape(ATH,PH1)
MH2 = mape(ATH,PH2)
MH12 = mape(PH1,PH2)
```

```python
print(f' MAPE ราคาทรัพย์กับผู้ประเมินคนที่ 1 คือ {MH1:.2f}\n MAPE ราคาทรัพย์กับผู้ประเมินคนที่ 2 คือ {MH2:.2f}\n MAPE ราคาตั้งขายของผู้ประเมินคนที่ 1 กับ ผู้ประเมินคนที่ 2 คือ {MH12:.2f}')
```
